## Creating the INSERT Statement

```python
insert_query = 'INSERT INTO predicted_outputs VALUES'
```

```python
insert_query
```

```python
df_new_obs.shape
```

```python
df_new_obs.columns.values
```

```python
df_new_obs['Age'].head()
```

```python
for i in range(df_new_obs.shape[0]): #We want to go through every row of df_new_obs
    insert_query += '(' #start with an open parenthesis
    
    for j in range(df_new_obs.shape[1]): #for every column of our dataframe, we take the value and enter with a comma
        insert_query += str(df_new_obs[df_new_obs.columns.values[j]][i]) + ', '
    
    insert_query = insert_query[:-2] + '),' #removed the last two characters and replace with closed parenthesis and comma
```

```python
insert_query
```

```python
insert_query = insert_query[:-1] +';'
```

```python
insert_query
```

```python
# Moves the data to MySQL Workbench
cursor.execute(insert_query)
```

```python
# we need to commit to add it to MySQL
conn.commit()
```

```python
# Close the connection
conn.close()
```
