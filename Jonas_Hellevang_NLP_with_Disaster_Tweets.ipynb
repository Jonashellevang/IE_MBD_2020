## Preprocessor

```python
import re,sys,string

def cleaned_text(text):
    text = text.lower()
    text = re.sub("\[.*?\]", "", text) # removing punctuation
    text = re.sub('https?://\S+|www\.\S+', '', text) #removing url
    text = re.sub('[()[\]{}\''',.``?:;!&^]','',text)
    text = re.sub('[%s]' % re.escape(string.punctuation), '', text)
    text = re.sub('\w*\d\w*', '', text)
    text = re.sub('û*','',text)
    text = re.sub('\n', '', text)
    text = re.sub('ûó*','',text)
    text = re.sub('Ûªs*','',text)
    text = re.sub('ò*','',text)
    text = re.sub(r'[^\w]', ' ', text)
    text = re.sub(r'^\d\S{0,}| \d\S{0,}| \d\S{0,}$', ' NUMBER ', text)
    text = re.sub(r"\s+", " ", text)
    text = re.sub(r"[0-9]", " ", text)
    text = re.sub(r"RT ", " ", text)
    text = re.sub(r"[^a-zA-Z#@\d\s]", " ", text)
    text = re.sub('@[^\s]+','',text)
    text = re.sub('#[^\s]+','',text)
    text = re.sub(r'\w+:\/{2}[\d\w-]+(\.[\d\w-]+)*(?:(?:\/[^\s/]*))*', '', text)
    text = re.sub(r"""
    (?:
      [<>]?
      [:;=8]                     # eyes
      [\-o\*\']?                 # optional nose
      [\)\]\(\[dDpP/\:\}\{@\|\\] # mouth      
      |
      [\)\]\(\[dDpP/\:\}\{@\|\\] # mouth
      [\-o\*\']?                 # optional nose
      [:;=8]                     # eyes
      [<>]?
    )""", "", text)
    
    return text
```

```python
train_df['clean_text'] = train_df['text'].apply(lambda x:cleaned_text(x))
test_df['clean_text'] = test_df['text'].apply(lambda x:cleaned_text(x))
```

## Lemmatizing with nltk

```python
def lemmatize_sentence(tokens):
    lemmatizer = nltk.stem.WordNetLemmatizer()
    lemmatized_sentence = []
    for word, tag in nltk.pos_tag(tokens):
        if tag.startswith("NN"):
            pos = "n"
        elif tag.startswith("VB"):
            pos = "v"
        else:
            pos = "a"
        lemmatized_sentence.append(lemmatizer.lemmatize(word, pos))
        lemmatized_sentence.append(" ")
    return "".join(lemmatized_sentence)
```

```python
train_df["clean_text"].str.split().apply(lemmatize_sentence)
test_df["clean_text"].str.split().apply(lemmatize_sentence)
```
