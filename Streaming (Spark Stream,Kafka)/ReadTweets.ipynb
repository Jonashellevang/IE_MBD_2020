```python

# Extract streaming data from Twitter, pre-processing
import access
import topic 
from pyspark import SparkConf, SparkContext
from pyspark.streaming import StreamingContext
from pyspark.sql import SparkSession
import re
import tweepy
import mysql.connector
import pandas as pd
from textblob import TextBlob
# Code connecting tweepy to MySQL adapted from multiple web sources and tweepy guide 


class MyStreamListener(tweepy.StreamListener):
    
    def on_status(self, status):
        
        # Extract tweets
        
        if status.retweeted:
            # Avoid retweeted info, and only original tweets will be received
            return True
        # Get useful fields from tweets
        id_str = status.id_str
        created_at = status.created_at
        text = deEmojify(status.text)    # Pre-processing the text  
        sentiment = TextBlob(text).sentiment
        polarity = sentiment.polarity
        subjectivity = sentiment.subjectivity
        user_created_at = status.user.created_at
        
        print(status.text)
        
        
        # Store all data in MySQL
        if Twitdb.is_connected():
            Twitcursor = Twitdb.cursor()
            sql = "INSERT INTO {} (id_str, created_at, text, polarity, subjectivity, user_created_at) VALUES (%s, %s, %s, %s, %s, %s)".format(topic.TABLE_NAME)
            val = (id_str, created_at, text, polarity, subjectivity, user_created_at)
            Twitcursor.execute(sql, val)
            Twitdb.commit()
            Twitcursor.close()
    
    
    def on_error(self, status_code):
       
        if status_code == 420:
            # return False to disconnect the stream
            return False
```

```python
def clean_tweet(self, tweet): 
    
    #clean tweets by removing special characters
    
    return ' '.join(re.sub("(@[A-Za-z0-9]+)|([^0-9A-Za-z \t]) \
                                |(\w+:\/\/\S+)", " ", tweet).split()) 
def deEmojify(text):
    
    #remove emojis 
    
    if text:
        return text.encode('ascii', 'ignore').decode('ascii')
    else:
        return None
```
